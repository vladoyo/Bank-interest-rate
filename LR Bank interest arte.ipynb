```python
#Analysys with multivariate linear regression model
import statsmodels.api as sm
dframe = pd.read_csv(r'C:/Users/Vlad/Desktop/ML course dataset/loan.csv')
```

```python
int_rate = dframe['Interest.Rate']
loan_amt = dframe["Loan.Amount"]
fico_score = dframe["FICO.Score"]
```

```python
loan_amt.shape
```

```python
fico_score.shape
```

```python
y = np.matrix(int_rate).transpose()
x1 = np.matrix(loan_amt).transpose()
x2 = np.matrix(fico_score).transpose()
```

```python
x = np.column_stack([x1, x2])
```

```python
x3 = sm.add_constant(x)
```

```python
x1.shape
```

```python
x2.shape
```

```python
x.shape
```

```python
x3.shape
```

```python
#Ordinay Least Square
model = sm.OLS(y, x3)
```

```python
model_fit = model.fit()
```

```python
print("The p-values are: ", model_fit.pvalues)
print("The R-squared are: ", model_fit.rsquared)
```
